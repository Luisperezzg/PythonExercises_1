# 1 Text Analysis Tool

```python
from collections import Counter
import re
from google.colab import files

def text_analysis(file_content):
    try:
        # Getting the contents from the uploaded file
        text = file_content.decode("utf-8")

        # Finding the number of paragraphs (separated by at least one newline character)
        paragraphs = [p for p in text.split('\n') if p]

        # Finding the number of sentences (ending with '.', '!' or '?')
        sentences = re.split(r'[.!?]', text)[:-1]

        # Finding the number of words and calculating the average word length
        words = re.findall(r'\b\w+\b', text)
        avg_word_length = sum(len(word) for word in words) / len(words)

        # Finding the most common words and their frequencies
        common_words = Counter(words).most_common(5)

        # Printing the results
        print(f'Paragraphs: {len(paragraphs)}')
        print(f'Sentences: {len(sentences)}')
        print(f'Words: {len(words)}')
        print(f'Avg word length: {avg_word_length:.2f}')
        print('Most common words:', ', '.join([f'{word[0]}({word[1]})' for word in common_words]))
    except Exception as e:
        # Handling any errors that occur during the file reading or analysis
        print(f'Error: {e}')

# Upload the file and get the file content
uploaded_files = files.upload()
for file_name in uploaded_files.keys():
    file_content = uploaded_files[file_name]
    text_analysis(file_content)
```

# 2 Hangman Game


```python
import random

# List of words for the hangman game
words = ["AMAZONS", "TIGRES", "BIOMEDICAL", "FIME", "SPINNING"]

# Welcome message and options to start or end the game
print("Welcome to the Hangman Game!")

# Randomly selects a word from the list
word_to_guess = random.choice(words)

# Stores the letters guessed by the player
guessed_letters = []

# Number of attempts allowed for the player
attempts = 4

# Helper function to display the current state of the word
def display_word():
    displayed_word = ""
    for letter in word_to_guess:
        if letter in guessed_letters:
            displayed_word += letter
        else:
            displayed_word += "_"
    return displayed_word

# Main loop for the hangman game
while attempts > 0:

    # Display the current state of the word and remaining attempts
    print(f"Word to guess: {display_word()}")
    print(f"Attempts remaining: {attempts}")

    # Ask the player for a new letter to guess
    guess = input("Please enter a letter to guess: ").upper()

    # Check if the letter has been guessed before
    if guess in guessed_letters:
        print("You have already guessed this letter.")
        continue

    # Add the guessed letter to the list of guessed letters
    guessed_letters.append(guess)

    # Check if the guessed letter is in the word
    if guess in word_to_guess:
        print("Good job! The letter is in the word.")

        # Check if the player has guessed all the letters
        if all(letter in guessed_letters for letter in word_to_guess):
            print(f"Congratulations! You've guessed the word: {word_to_guess}")
            break
    else:
        print("The letter is not in the word.")
        attempts -= 1
else:
    # If the loop completes without breaking, the player has run out of attempts
    print(f"Sorry, you've run out of attempts. The word was: {word_to_guess}")D
```

# 3 Number Guessing Game

```python
import random

# Generate a random number between 1 and 100
target_number = random.randint(1, 100)

# Initial message to the user
print("I have generated a random number between 1 and 100. Try to guess it!")

# Loop to keep asking the user for a guess until they guess it correctly
while True:
    # Get the user's guess as an integer
    user_guess = int(input("Please enter your guess (between 1 and 100): "))

    # Check if the guess is less than 1 or greater than 100 and ask for a valid guess if necessary
    if user_guess < 1 or user_guess > 100:
        print("Invalid guess. Please enter a number between 1 and 100.")
        continue

    # Check if the user's guess is less than, equal to, or greater than the target number
    if user_guess < target_number:
        print("Too low!, the number is greater than this one. Please,Try again.")
    elif user_guess > target_number:
        print("Too high!,  the number is lower than this one. Please,Try again.")
    else:
        print("Congratulations! You've guessed the correct number.")
        print(f"You guessed the number: {target_number}")  # Corrected line
        break  # Exit the loop as the user has guessed correctly

```
